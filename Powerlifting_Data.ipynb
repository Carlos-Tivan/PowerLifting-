```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('powerlifting_dataset.csv') #import data
data.head(5)
```

```python
# Rename data coulumns Lifter Name , Age weight , Lift Type and Kg

data_rename = data[['Lifter Name','Age','Weight Class','Lift Type','Amount Lifted (kg)']]
data_rename.head(5)

# Rename columns
data_rename.columns = ['Name','Age','Weight','Lift Type ','Lifted_kg']
data_rename.head(5)
```

```python
# Show unique values for 'Age' and 'Lift Type'

# Get the unique values from the 'Age' column
age = data.Age.unique()

# Get the unique values from the 'Lift Type' column
lift_type = data.Lift_Type.unique()

# Print the unique values of 'Age' and 'Lift Type'
print(age)
print(lift_type)


```

```python
# Sort Age	Weight Class

data_sort = data_rename.sort_values(by=['Age','Weight'])
data_sort.head(5)
```

```python


# Define the age range (for example, ages between 18 and 30)
age_range = data_rename[(data_rename['Age'] >= 18) & (data_rename['Age'] <= 30)]

# Sort the filtered DataFrame by 'Age' and 'Lifted_kg'
data_sort = age_range.sort_values(by=['Age', 'Lifted_kg'], ascending=True)

# Display the first 5 rows of the sorted DataFrame
print(data_sort.head(5))

print('\n')

# Display the last 5 rows of the sorted DataFrame
print(data_sort.tail(5))

```

```python
# Plotting histograms with seaborn

# Histogram for 'Age'
plt.figure(figsize=(12, 6))  # Create a new figure with a specified size
sns.histplot(data_sort['Age'], bins=10, color='skyblue', kde=False)  # Create a histogram for the 'Age' column
# `bins=10` specifies the number of bins for the histogram.
# `color='skyblue'` sets the color of the bars in the histogram.
# `kde=False` disables the Kernel Density Estimate (KDE) curve, showing only the histogram.

plt.title('Histogram of Age')  # Set the title of the histogram
plt.xlabel('Age')  # Set the label for the x-axis
plt.ylabel('Frequency')  # Set the label for the y-axis
plt.grid(True)  # Add grid lines to the plot for better readability
plt.show()  # Display the histogram

# Histogram for 'Lifted_kg'
plt.figure(figsize=(12, 6))  # Create a new figure with a specified size
sns.histplot(data_sort['Lifted_kg'], bins=10, color='salmon', kde=False)  # Create a histogram for the 'Lifted_kg' column
# `bins=10` specifies the number of bins for the histogram.
# `color='salmon'` sets the color of the bars in the histogram.
# `kde=False` disables the Kernel Density Estimate (KDE) curve, showing only the histogram.

plt.title('Histogram of Lifted Kg')  # Set the title of the histogram
plt.xlabel('Lifted Kg')  # Set the label for the x-axis
plt.ylabel('Frequency')  # Set the label for the y-axis
plt.grid(True)  # Add grid lines to the plot for better readability
plt.show()  # Display the histogram
```
